```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
test_X = np.array([[0.5, 2],
                   [2, 2],
                   [-0.5, 0],
                   [-0.5, -2]])
test_y = np.array([1, -1, 1, -1])

plt.figure()
for x, y in zip(test_X, test_y):
    plt.scatter(x[0], x[1], marker = "o" if y == 1 else "x", color = "black")

x_vals = np.linspace(-2.5, 2.5, 100)
y_vals = x_vals + 1
plt.plot(x_vals, y_vals)

plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)

plt.show()
```

```python
train_X = np.array([[1, 1],
                    [-1, 1],
                    [-1, -1],
                    [1, -1]])
train_y = np.array([-1, 1, 1, -1])

all_X = np.vstack([train_X, test_X])
all_y = np.concatenate([train_y, test_y])

plt.figure()
for x, y in zip(all_X, all_y):
    plt.scatter(x[0], x[1], marker = "o" if y == 1 else "x", color = "black")

x_vals = np.linspace(-2.5, 2.5, 100)
initial_y_vals = -x_vals + 1
final_y_vals = x_vals + 1
optima_y_vals = 2 * x_vals

plt.plot(x_vals, initial_y_vals, label = "Initial")
plt.plot(x_vals, final_y_vals, label = "Final")
plt.plot(x_vals, optima_y_vals, linestyle = "--", label = "Optimal")

plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 2.5)

plt.legend(loc = "lower right")
plt.show()
```
